```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import os
```

```python
def read_json(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    return data
```

```python
score_dir = 'base_sgd_sh_checkpoint/metrics/'
score_json_lst = [score for score in os.listdir(score_dir) if score.endswith('.json')]
score_dict_lst = []
for i, score in enumerate(score_json_lst):
    score_dict_lst.append(read_json(os.path.join(score_dir, score_json_lst[i])))

score_dict_lst
```

```python
data_sorted = sorted(score_dict_lst, key=lambda x: x['epoch'])

# epoch, train_loss, validation_mpjpe 추출
epochs = [item['epoch'] for item in data_sorted]
train_losses = [item['train_loss'] for item in data_sorted]
validation_mpjpes = [item['validation_mpjpe'] for item in data_sorted]
```

```python
# 첫 번째 그래프: Train Loss
plt.figure(figsize=(10, 5))
plt.plot(epochs, train_losses, label='Train Loss', marker='o', color='b')
plt.title('Train Loss over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Train Loss')
plt.grid(True)
plt.legend()
plt.show()

# 두 번째 그래프: Validation MPJPE
plt.figure(figsize=(10, 5))
plt.plot(epochs, validation_mpjpes, label='Validation MPJPE', marker='x', color='r')
plt.title('Validation MPJPE over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Validation MPJPE')
plt.grid(True)
plt.legend()
plt.show()
```
